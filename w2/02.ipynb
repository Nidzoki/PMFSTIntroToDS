#### **Zadatak 2.**
Učitajte skup podataka `tips` u DataFrame s nazivom varijable `tips_df`.
Skup podataka sadrži podatke o narudžbama u jednom restoranu. Pogledajte kakvi su podaci u tom skupu podataka:

* koje tipove podataka sadrži skup podataka?

```python
import pandas as pd
tips_df = sns.load_dataset("tips")
print(" Dataset ima: - ", tips_df.shape[0], "redaka\n\t     - ", tips_df.shape[1], " stupaca\n")
print(" Sadrži ukupno: \n", tips_df.size, " podataka.\n")
print(" Nazivi stupaca: \n", tips_df.columns)
print("\n Tipovi:\n", tips_df.dtypes)
print("\n Prva četiri zapisa: \n", tips_df.head(4))
print("\n Zadnjih 6 zapisa: \n", tips_df.tail(6))
```

#### **Zadatak 3.**

#### **Zadatak 7.**
Prikažite `scatterplot` gdje su na x-osi vrijednosti iz `total_bill` stupca, a na y-osi vrijednosti iz `tip` stupca. Obojajte "točkice" ovisno o stupcu `size`.

> Hint: [scatterplot dokumentacija](https://seaborn.pydata.org/generated/seaborn.scatterplot.html?highlight=scatterplot#seaborn.scatterplot)

Razmislite što vizualizacija pokazuje te može li se nešto zaključiti iz vizualizacije.

```python
tips_df.plot.scatter(x="total_bill", y="tip", c="size", colormap="rainbow")
print("Što je viši total_bill, viši je i tip.\n")
```

#### **Zadatak 8.**
Uredite prethodnu vizualizaciju tako da:
* dodate naslov vizualizacije
* uredite opise za x i y os
* promijenite boje "točkica"
* uredite legendu (promijenite naslov, lokaciju, ...)
* dodate više ticks na y-osi (npr. da umjesto 0, 10, 20,... budu 0, 5, 10, 15, 20, ...)
* ...

**Napomena:** Seaborn je neka vrsta nadogradnje Matplotlib-a - mogu se koristiti iste/slične naredbe.

```python
p = sns.scatterplot(x="total_bill", y="tip", hue="size", palette="rainbow", data=tips_df, legend='full')
p.set_title("Ovisnost napojnice o iznosu računa")
p.set_xlabel("Iznos računa")
p.set_ylabel("Napojnica")
p.set_yticks([0, 5, 10, 15, 20])
p.legend(title="Veličina grupe")
```
